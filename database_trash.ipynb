```python
import pickle, sys
sys.path.append('./database')
from persons import Person, Article, DataBase, correct_str
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import wikipediaapi
options = webdriver.ChromeOptions()
options.add_argument("--headless")
driver = webdriver.Chrome('database/chromedriver.exe', options=options)
```

```python
with open('./database/database.pickle', 'rb') as f:
    orig_database = pickle.load(f)
with open('./database/database_trash.pickle', 'rb') as f:
    database = pickle.load(f)
# database = DataBase()
# with open('./database/database_trash.pickle', 'wb') as f:
    # pickle.dump(database, f)
```

```python
from numpy import random
from time import sleep
import nltk
nltk.download('punkt')
```

```python
while True:
    driver.get('https://en.wikipedia.org/wiki/Special:RandomInCategory')
    search_box = driver.find_element_by_name('wpcategory')
    search_box.send_keys('Living people')
    search_box.submit()

    url = driver.current_url
    response = requests.get(url=url)
    soup = BeautifulSoup(response.content, 'html.parser')
    title = soup.find(id="firstHeading").text
    while title[-1] == ' ' or title[-1] == '\t' or title[-1] == '\n':
        title = title[:-1]
    print(title)
    corr_title = correct_str(title)
    while corr_title[-1] == ' ' or corr_title[-1] == '\t' or corr_title[-1] == '\n':
        corr_title = corr_title[:-1]

    wiki = wikipediaapi.Wikipedia('en')
    page = wiki.page(title)
    summary = page.summary
    if (
        'football' in summary or 'baseball' in summary or 'volleyball' in summary 
        or len(nltk.tokenize.sent_tokenize(summary)) < 5):
        print('Bad person')
        continue

    if (
        orig_database.find(corr_title) == -1 and
        database.find(corr_title) == -1):
        res = database.add([corr_title, title], None, wikitext=summary)
        if res < 2:
            database.remove(corr_title)
        with open('./database/database_trash.pickle', 'wb') as f:
            pickle.dump(database, f)

    print("Database:", len(database.persons))
    sleeptime = random.uniform(70, 80)
    sleep(sleeptime)
    print('\n')
```
